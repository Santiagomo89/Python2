```python
class producto:
    def __init__(self, codigo, nombre, precio_producto, descuento):
        self.codigo = codigo
        self.nombre = nombre
        self.precio_producto = precio_producto
        self.descuento = descuento
    #métodos personalizados
    def codigonombre(self):
        print("El código + nombre es",self.codigo + "" +self.nombre)
    def precio_neto(self):
        resultado = self.precio_producto * (1-self.descuento)    
        return resultado
```

Una vez que tengamos esto, necesitamos tener: 
- un método que muestre la marca y el modelo separado por una barra /
- un método que mire si el año del vehículo es aneterior a 2023, y que muestre con un texto "Este vehículo tiene que pasar la ITV"
- un método que, si la etiqueta ECO = B, incremente el precio del vehículo en un 18.5%, y que si la etiqueta ECO = C, lo incremente en un 40.6%. He de poder usar el precio final en una operación aritmética matemática
- un método que le reste un 12% a la Velocidad Max para que sea = velocidad óptima

```python
class vehiculo:
    def __init__(self, Marca, Modelo, Año, Velocidad_Max, Precio, Etiqueta_ECO, Tasa_ECO):
        self.Marca = Marca
        self.Modelo = Modelo
        self.Año = Año
        self.Velocidad_Max = Velocidad_Max
        self.Precio = Precio
        self.Etiqueta_ECO = Etiqueta_ECO
        self.Tasa_ECO = Tasa_ECO

    def separbarra(self):
        print(self.Marca + "/" + self.Modelo)

    def pre2023(self):
        if self.Año < 2023:
            print("Este vehículo tiene que pasar la ITV")

    def todosomoshacienda(self):
        if self.Etiqueta_ECO == "B":
            return (self.Precio*1.185)
        elif self.Etiqueta_ECO == "C": 
            return (self.Precio*1.406)
        else:
            return self.Precio   

    def velocidadoptima(self):
        print("La velocidad óptima (el 88% de la velocidad máxima) es",(self.Velocidad_Max*0.88))    
```

```python
from datetime import date as fecha

def diastransc(fechainput):
    hoy = fecha.today()
    fechacreada = fechainput.split("@")
    #fechacreada será una lista que contenga tres elementos, el primero será indice 0, el segundo 1, y el tercero 2
    dia = int(fechacreada[0])
    mes = int(fechacreada[1])  
    año = int(fechacreada[2])
    fechafinal = fecha(day=dia, month=mes, year=año)
    #restamos las 2 fechas
    print("Los días transcurridos serían un total de", (fechafinal - hoy).days)

diastransc(input("Inserte aquí una fecha separada por arrobas en formato de día, mes año, p.ej 15@06@2025"))   
```
