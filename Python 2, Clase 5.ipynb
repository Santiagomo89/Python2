Por lo general, el DataFrame genérico siempre es llamado df por convención de programación

```python
#Cómo añadir una columna a un DataFrame
#Opción 1 - Poniendo un valor fijo (para todas las filas)
#opción 2 - Con el método Apply (genérico/específico)
#Opción 3 - Usando una o varias columnas existentes
#Opción 4 - Con los métodos loc e iloc (me permite hacer lectura/escritura de un dato concreto de una celda)
```

```python
#Opción 1. Si no existe el nombre de la columna, te la crea nueva
pedidos["Comercial"] = "Manolo Sánchez"
pedidos
```

```python
#Opción 2. Método apply, necesitamos una función. Si la cantidad es mayor de 30, le pongo en una columna Dto = Sí, y si no Dto = No
def calcular_dto(cantidad):
    if cantidad>30:
        return "Sí"
    else:
        return "No"
#Aplicar la función para crear una columna nueva
pedidos["Dto"] = pedidos["Cantidad"].apply(calcular_dto)   
pedidos 
```

```python
#Opción 2. Apply genérico, en el que no llamo a la función desde una columna, si no desde el dataframe entero; por lo tanto, mi función cambia
def calcular_dto_gen(fila):
    return fila["Cantidad"] * fila["Precio Ud"]
#Llamar a apply genérico, hay que indicarle el eje (axis=1, columnas)
pedidos["Importe"] = pedidos.apply(calcular_dto_gen, axis=1)
pedidos
```

```python
#Opción 3. Usando una o varias columnas existentes (operación vectorizada)
pedidos["Importe2"] = pedidos["Precio Ud"] * pedidos["Cantidad"]
pedidos
```

```python
#Opción 3. Una sola columna, el importe de descuento será del 25% sobre el importe
pedidos["Importe Descuento"] = pedidos["Importe"] * 0.25
pedidos
```

```python
#Opción 4. Mediante iloc (INTEGER LOCATION), es de lectura/escritura
pedidos.iloc[0,0] #el valor de la primera columna y de la primera fila
pedidos.iloc[0,0] = "Nuevo cliente"
pedidos
#Y así es como lo escribimos
```

PRÁCTICA 26: Aprovechando el dataframe de pedidos, crear la columna "Precio Unidad 2", que incremente el Precio Unidad en un 15%

```python
pedidos["Precio Unidad 2"] = round(pedidos["Precio Ud"] * 1.15,2) #El round no es necesario, lo ponemos tan solo por estetica
pedidos
```

PRÁCTICA 27: Aprovechando el dataframe de pedidos, crear la columna "Cantidad 3" que incremente la cantidad en un 25% si la cantidad es mayor de 45

```python
def calcular_can(cantidad):
    if cantidad>45:
        return cantidad * 1.25
    else:
        return cantidad
#Aplicar la función para crear una columna nueva
pedidos["Cantidad 3"] = pedidos["Cantidad"].apply(calcular_can)   
pedidos 
```

PRÁCTICA 28: Aprovechando el dataframe de pedidos, crear la columna "Diferencia Cantidad", que me de la diferencia entre la cantidad más grande y la cantidad de cada fila

```python
maxim = pedidos["Cantidad"].max()

def calcular_difcan(fila):
    return maxim - fila["Cantidad"]

pedidos["Diferencia Cantidad"] = pedidos.apply(calcular_difcan, axis=1)
pedidos
```
