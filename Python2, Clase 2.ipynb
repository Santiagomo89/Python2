La estructura MATCH... CASE. En python solo está disponible a partir de la versión 3.10 y solo sirve para valores exactos

```python
x = 25
match x:
    case 10:
        print("El número es 10")
    case 20:
        print("El número es 20")
    case 30:
        print("El número es 30")    
    case default:
        print("Otro número distinto del 10, 20 o 30")        
```

Un bucle es un código que se repite hasta que se cumple una condición dada. Tenemos dos tipos de buclles, finitos e infinitos.
- Los finitos son aquellos que van a terminar sí o sí, ya que conocemos de antemano el número de veces que se repetirá el bucle. Cada repetición de bucle se denomina iteración... así que si sabes las iteraciones, lo tienes. Los bucles infinitos tienhen una variable asociada
- Los infinitos puede que terminen, o puede que no

```python
#bucle finito con for
for i in [1,2,3,4,5]:
    print("El valor de i es",i)
print("Terminado")    
```

```python
for i in [25, 36, 97]:
    print("El valor de i es",i)
```

```python
#Lo recomendable es tener nuestro contador iter
iter = 0
listanum = [25, 36, 14, 27, 81, 96]
for i in listanum:
    print("El número es",i)
    #iter=iter+1 Opción A (amateur)
    iter+=1 #Opción B (la opción pro) 
print("El número de iteraciones es",iter)    
```

Las FUNCIONES PERSONALIDAS en python.

Una función es un código que devuelve un resultado. Una función empieza por la palabra clave DEF, después un espacio en blanco, y a continuación el nombre de la función; este nombre no puede contener espacios en blanco. Después, presenta un paréntesis de apertura y dentro de este irían los parámetros, que son los valores que quieres pasarle a la función, cerrados por un paréntesis de cierre.
- El nombre de la función no puede ser un nombre reservado (que ya exista, como RANGE o FOR), y luego dos puntos :
- No puede contener espacios en blanco
- No puede contener caracteres especiales
- No puede empezar por un número
- Si hay +1 parámetro, separado por cosas
La función debe ejecutarse primero y después llamarse, cuando se ejecuta se guarda en el kernel. Si se modifica la función se tiene que volver a ejecutar.

```python
#Una función que le sume 3 a un número, primero la creo
def suma3(x):
    #Cómo devuelve la función el valor
    print("El resultado es",x+3)
suma3(10) #y aquí la llamo    
```

```python
#Una función que le sume 3 a un número, primero la creo
def suma3(x):
    #Cómo devuelve la función el valor
    print(x+3)
suma3(10)+22 #y aquí la llamamos otra vez, pero da fallo porque lo que queremos es una operación aritmética, y PRINT da un texto   
```

```python
#En su lugar...

def suma3(x):
    return x+3
suma3(10)+22

#El return devuelve el valor correcto
```

```python
def restaab(a,b):
    return a-b
#la llamamos
restaab(10,3)

#Los parámetros siguen el orden por defecto, pero esto puede cambiar
restaab(b=15, a=30)
```

Las CLASES PERSONALIZADAS en python. Una clase es como una plantilla que se puede usar para crear objetos y que herede sus atributos y métodos. Una clase tiene tres elementos principales:
- Métodos especiales (_nombre_)
- Métodos personalizados, los que se crea uno mismo
- Los atributos, propiedades
Una clase se crea o define con la palabra reservada class

```python
class alumno: 
    #vamos a iniciar la clase con el método especial _init_
    def _init_(self, nombre, edad, estatura): #El parámetro self se pone siempre cuando se usa una clase, y se refiere a si misma
        self.nombre = nombre
        self.edad = edad
        self.estatura = estatura #los atributos dentro de la clase
#Vamos a crear dos métodos personalizados
    def saludo(self):
        print("Hola, soy la IA,",self.nombre, "te doy la bienvenida.")
    def restaltura (self):
        print("La altura menos 10 cms es:",self.estatura-10)
    def animos(self):
        print(self.nombre, "hoy el jefe te sube el sueldo.")        
```

```python
#creamos una instancia de la clase alumno
alumno1 = alumno("Juan Carlos", 44, 1.80)
alumno2 = alumno("María", 36, 1.67)

alumno1.saludo()
alumno2.animos()
alumno1.restaltura()
```
