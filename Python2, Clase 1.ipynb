¿Qué es una LISTA? Una colección ordenada de objetos, que es mutable y permite guardar varios valores en una variable como un vector
Los elementos de una lista se separan por comas
Los elementos de una lista van indexados, y empiezan siempre por 0
Puedes crear una lista vacía para, por ejemplo, ir llenándola con un bucle. Una lista vacía sería nombre_lista = []
Para ver el numero de elementos de una lista, usaremos len

```python
compra = ["Pan", "Leche", "Arroz", "Nutella", 10, 25, 35.6, True]
print(compra)
len(compra)
```

```python
print(compra[1])
print(compra[3])
```

```python
#Como añadir un elemento al final de una lista, con el método append
compra.append(52)
compra
#Cuidado, si le das más de una vez, saldrá varias
```

```python
#Como insertar un elemento en una lista, con el método insert, donde ponemos dos datos: la posición y su información
compra.insert(1, "Pescado")
compra
```

```python
#Como añadir más de un elemento a la vez, con extend, a una lista
compra2 = ["Lentejas", "Amborgesas"]
compra.extend(compra2)
compra
```

```python
#Como eliminar un elemento de una lista, con el método remove o con la función del. REMOVE pones el valor del elemento y elimina el primero que encuentra, DEL puede 
#indicar el valor a eliminar. Para eliminar TODOS LOS VALORES DE UNA LISTA, usamos el método CLEAR
compra.remove(52)
compra
```

El DICCIONARIO es una colección de pares CLAVE + VALOR, sin orden fijo por índice. Tip importante, la clave debe ser única, se crea con la llave {}.
Los diccionarios son mutables (son de escritura). dic = {} es un modelo de diccionario vacío

```python
alumno = {
    "Nombre": "Andrea",
    "Edad": 34,
    "Nota": 8.5,
}
print(alumno["Nombre"])
print(alumno["Nota"])

#¿Como evitar errores si la clave no existe? Usando el método get interceptamos el error y damos tratamiento
print(alumno.get("Materia", "No existe dicha clave"))

#Ver sólo las claves, con keys
print(alumno.keys())

#Ver sólo los valores
print(alumno.values())
```

ESTRUCTURAS DE DECISIÓN, el Condicional. IF... ELIF... ELSE, SWITCH... CASE
- El if analiza una condición o prueba lógica y si se cumple la condición ejecuta un código; si no se cumple, puede ejecutar otro código
Prueba lógica = comparar valores mediante operadores de comparación
- > Mayor que (solo aplica a números)
- < Menor que (solo aplica a números)
- >= Mayor o igual que (solo aplica a números)
- <= Menor que (solo aplica a números)
- == Igual que
- != Distinto a
En python toda estructura empieza con una linea con dos puntos al final. Siempre que hay :, la siguiente perteneciente a la estructura va indentada 4 espacios a la derecha

```python
i = 17410
if i<100:
    print("El número",i,"es menor que 100")
else:
    print("El número",i,"no es menor que 100")    
```

```python
#Más de dos opciones de devolución
i = 0
if i<50:
    print("El valor es menor que 50")
elif i<100:   
    print("El valor está entre 50 y 100") 
else:
    print("El valor emayor o igual que 100")    
```

Las funciones de lógica OR y AND
- OR devuelve verdadero si se cumple UNA SOLA de las condiciones que se analizan
- AND devuelve verdadero si se cumplen TODAS las condiciones que se analizan

```python
i = 50
if i>=10 and i<=50:
    print("Dentro del tramo 10-50, ambos inclusive")
```

```python
#Números positivos y menores de 100
i = 30
if i<0 or i>=100:
    print("Fuera del tramo, ya sea negativo o ser cien (o mayor)")
```
