PRACTICA 1: Vamos a crear una lista llamada meses, donde tendremos los meses de enero a julio menos, añadiremos a esta lista el mes de agosto, insertaremos en su lugar el mes de abril, añadire de una sola vez los meses que me faltan hasta diciembre y después. Después, añadiré al final dos veces mayo. Finalmente, eleminaré los meses incorrectos, e iré mostrando como evoluciona la lista

```python
# Solución a la práctica 1
meses = ["enero", "febrero", "marzo", "mayo", "junio", "julio"]
print(meses)
meses.append("agosto")
print(meses)
meses.insert(3, "abril")
print(meses)
meses2 = ["septiembre", "octubre", "noviembre", "diciembre"]
meses.extend(meses2)
print(meses)
meses.append("mayo")
meses.append("mayo")
print(meses)
while meses[-1] == "mayo":
    del meses[-1]
print(meses)
```

# Estructura de decisión

#### El condicional if, elif, else, switch, case

el IF analiza una condición o prueba lógica y si se cumple la condición ejecuta el código, si no se cumple puede ejecutar otro código.
- Prueba lógica = comparar valores mediante operadores de comparación
- (>) Mayor que (solo a números)
- < Menor que (solo a números)
- (>=) Mayor o igual que (solo números)
- <= Menor o igual que (solo números)
- == Igual que
- =! Distinto que
- En python toda estructura empieza con una linea con dos puntos al final siempre que hay: la siguiente linea pertenece a la estructura va indentada 4 espacios a la derecha

```python
i = 50
if i < 100:
    print(f"El número {i} es menor que 100")
else:
    print(f"El número {i} es mayor que 100")
print("Terminado")
```

```python
# Más de dos opciones de devolución

i = 100

if i < 50:
    print(f"El valor {i} es menor que 50")
elif i < 100:
    print(f"El valor {i} está entre 50 y 100")
else:
    print(f"El valor {i} es mayor o igual a 100")
```

Funciones de lógica OR y AND
- OR devuelve verdadero si se cumple una sola de las condiciones que se analizan
- AND devuelve verdadero si se cumplen todas las condiciones que se analizan

```python
i = 75

if i > 10 and i < 50:
    print(f"El valor {i} está dentro del tramo 10-50")
```

```python
# Números positivos y menores de 100
i = 130

if i < 0 or i > 100:
    print(f"El número {i} está fuera del tramo")
```

PRÁCTICA 3: Mediante un input poniendo un número del 1 al 12 más una estructura IF, devolver el nombre del mes en letras, mostrar un mensaje en caso que el número correcto (solo es correcto del 1 al 12)

```python
user_input = int(input("Ingresar un número del 1 al 12"))

if user_input == 1:
    print(f"El mes correspondiente al número {user_input} es Enero")
elif user_input == 2:
    print(f"El mes correspondiente al número {user_input} es Febrero")
elif user_input == 3:
    print(f"El mes correspondiente al número {user_input} es Marzo")
elif user_input == 4:
    print(f"El mes correspondiente al número {user_input} es Abril")
elif user_input == 5:
    print(f"El mes correspondiente al número {user_input} es Mayo")
elif user_input == 6:
    print(f"El mes correspondiente al número {user_input} es Junio")
elif user_input == 7:
    print(f"El mes correspondiente al número {user_input} es Julio")
elif user_input == 8:
    print(f"El mes correspondiente al número {user_input} es Agosto")
elif user_input == 9:
    print(f"El mes correspondiente al número {user_input} es Septiembre")
elif user_input == 10:
    print(f"El mes correspondiente al número {user_input} es Octubre")
elif user_input == 11:
    print(f"El mes correspondiente al número {user_input} es Noviembre")
elif user_input == 12:
    print(f"El mes correspondiente al número {user_input} es Diciembre")
else:
    print(f"Número {user_input} fuera de rango")
```
