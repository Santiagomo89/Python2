```python
from datetime import date as fechas
#Vamos a crear la instancia "hoy"
hoy = fechas.today()
print(hoy)

#Ajustar el idioma para ver los días en espanish
import locale
locale.setlocale(locale.LC_TIME, "es_ES.UTF-8")

#Cómo dar formato a las fechas de la clase date, con strftime
print("El día de hoy es",hoy.strftime("%d")) #%d es el día en número, pero es texto porque sale de un str
print("El día de hoy en letras es",hoy.strftime("%A")) #el dína en letra (también texto, sigue siendo un str)
print("El día de hoy en letras es",hoy.strftime("%A").upper()) #el dína en letra (también texto, sigue siendo un str) #lo mismo, pero para ver que se puede poner
print("El día de hoy en número de la semana es",hoy.strftime("%w")) #esto indiva la posición numeral del día de la semana, sigue siendo un str, ya sabes
print("El día del año en número es",hoy.strftime("%W")) #esto te indica qué semana del año sería, en este caso la 12ava, en este caso la W mayúscula
print("El mes de hoy en letras es",hoy.strftime("%B")) #esto te dice el mes entero en letras
print("El mes de hoy en letras pero abreviado es",hoy.strftime("%b")) #lo mismo pero abreviado
print("El año de hoy es",hoy.strftime("%Y")) #año en números

#Cómo personalizar una fecha a partir de una instancia date
print("La fecha de hoy es",hoy.strftime("%d/%m/%y"))
```

```python
#Serie numérica
lista_numeros = [15,26,33,24,75,89,78,96,100]
snumeros = pd.Series(lista_numeros)
snumeros.info()
```

```python
#Funciones de resumen que podemos usar sobne una serie numérica
print("La suma de los valores de la serie es",snumeros.sum())
print("El promedio de los valores de la serie es",snumeros.mean())
#Redondeado
print("El promedio de los valores de la serie es",round(snumeros.mean(), 2))
print("El valor más grande de la serie es",snumeros.max())
print("El menor de los valores de la serie es",snumeros.min())
print("El número de valores de la serie es",snumeros.count())
```

```python
#Métodos estadísticos sobre series numéricas
print("La suma acumulada es",snumeros.cumsum())
#La varianza mide cuánto se separan los datos respecto a su media aritmética
print("El varianza es",snumeros.var())
#La desviación típica es la raíz cuadrada de la varianza, me devuelve las unidades originales
print("La desviación típica es es",snumeros.std())
```

```python
#Cómo transformar los datos de una serie numérica
snumeros*2 #multiplicamos por 2 cada elemento de la serie. Si quisiéramos cambiarla a futuro con todas las cosas, ya que esto no modifica la serie original, lo que tendríamos
#que hacer es hacer snumeros = snumeros*2, lo cual si lo modificaría
```

```python
#Dividir por tres, pero no se modifica el original, de nuevo (ver arriba)
snumeros/3
```

```python
#Ordenación + filtrado
snumeros.sort_values(ascending=True) #primero la ordenación
snumeros[snumeros>50] #y ahora el filtrado, después del nombre de la serie, ponemos corchetes y dentro pondremos el filtro
snumeros[snumeros>50].sort_values(ascending=True) #y con esto, además, podríamos combinar ambos
```

PRÁCTICA 25: Partiendo de la lista VIP = [18,35,78,44,63,24,27,88,125,7,53] la convertiremos en una serie de pandas, haremos una copia de la serie solamente para los valores menores de 70 (VIP70) y la queremos ver en orden descendente. Aparte de esto, queremos: 
- Ver en un texto la suma de los valores de la serie VIP original
- Ver en un texto el promedio de la serie VIP70
- Ver en un texto la desviación típica de la serie VIP original

```python
import pandas as pd

VIPoriginal = pd.Series([18,35,78,44,63,24,27,88,125,7,53])

VIP70 = VIPoriginal[VIPoriginal<70]
VIP70.sort_values(ascending=False)
```

```python
print("La suma de los valores de la serie original es",VIPoriginal.sum())
print("El promedio de los valores de la serie VIP70 es",VIP70.mean())
print("La desviación típica dela serie VIP original es",round(VIPoriginal.std(), 3))
```
