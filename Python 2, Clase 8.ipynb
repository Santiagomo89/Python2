PRÁCTICA 37: Conectarse a tu servidor de Base de datos SQL Server Express y traer en 2 dataframes distintos, por un lado la tabla Pedidos y por otro lado la tabla de detalles de pedidos
(calculando en la sentencia sql el importe = Precio unidad x Cantidad) con los 2 dataframes hacer merge y obtener la suma de importe por CiudadDestinatario

```python
import pandas as pd
import pyodbc
Cadena = (
    f'Driver={{ODBC Driver 17 for SQL Server}};'
    f'Server={r'M027599\SQLEXPRESS'};' #Este lo copiamos dando click derecho al server que tenemos en el programa de SQL y copiando lo que vemos en Propiedades
    f'Database={'Neptuno'};'
    f'Trusted_Connection={'yes'};'
)

Conexion = pyodbc.connect(Cadena)
```

```python
pedidos = pd.read_sql("Select * from Pedidos", Conexion)
pedidos
```

```python
detallespedidos = pd.read_sql("Select IdPedido, PrecioUnidad, Cantidad, (PrecioUnidad * Cantidad) as Importe from Detalles_de_pedidos", Conexion)
detallespedidos
```

```python
fusion = pd.merge(pedidos, detallespedidos, on="IdPedido", how="inner")
fusion
```

```python

resultado = fusion.groupby(["CiudadDestinatario"]).Importe.sum()
resultado = resultado.reset_index() #Esto no es necesario, es solo por estética de dataframe en formato pandas
resultado
```

```python
#copiamos el enlace que hemos sacado del ssms, borrando los espacios
sentencia = "SELECT Pedidos.CiudadDestinatario, SUM(Detalles_de_pedidos.PrecioUnidad * Detalles_de_pedidos.Cantidad) AS Importe FROM Pedidos INNER JOIN Detalles_de_pedidos ON Pedidos.IdPedido = Detalles_de_pedidos.IdPedido GROUP BY Pedidos.CiudadDestinatario"
```

```python
resultado = pd.read_sql(sentencia, Conexion)
resultado
```

Cómo tratar una sentencia SQL que viene del Asistente de Vistas. Usaremos para esto la libreria sqlparse, que se tiene que instalar (>pip install sqlparse) - HECHO

```python
#Llamar a la librería sqlparse
import sqlparse
#copiar la sentencia original, con sus espacios mal puestos y todo, del original
codeoriginal = """SELECT        Pedidos.CiudadDestinatario, SUM(Detalles_de_pedidos.PrecioUnidad * Detalles_de_pedidos.Cantidad) AS Importe
FROM            Pedidos INNER JOIN
                         Detalles_de_pedidos ON Pedidos.IdPedido = Detalles_de_pedidos.IdPedido
GROUP BY Pedidos.CiudadDestinatario""" #Triple comilla para hacer esto, no comilla a secas
codeformateado = sqlparse.format(
    codeoriginal, 
    reindent=True, 
    keyword_case="upper"
)

print(codeformateado)
```

```python
agrupado = pd.read_sql(codeformateado, Conexion)
agrupado
```

PRÁCTICA 38: De los datos del SQL Server, necesito guardar en un fichero de excel la suma de importe, más el número de pedidos, por nombre de categoría

```python
import sqlparse
codesql = """SELECT        SUM(Detalles_de_pedidos.PrecioUnidad * Detalles_de_pedidos.Cantidad) AS Importe, Categorías.NombreCategoría AS NumPedidos, COUNT(Pedidos.PaísDestinatario) AS Expr1
FROM            Pedidos INNER JOIN
                         Detalles_de_pedidos ON Pedidos.IdPedido = Detalles_de_pedidos.IdPedido INNER JOIN
                         Productos ON Detalles_de_pedidos.IdProducto = Productos.IdProducto INNER JOIN
                         Categorías ON Productos.IdCategoría = Categorías.IdCategoría
GROUP BY Categorías.NombreCategoría"""

codelimpio = sqlparse.format(
    codesql, 
    reindent=True, 
    keyword_case="upper"
)

resultado = print(codelimpio)

agrupado = pd.read_sql(codelimpio, Conexion)
agrupado
```
