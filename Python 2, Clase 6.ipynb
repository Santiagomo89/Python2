PRÄCTICA 33: Mediante el dataframe trimestre, averiguar la suma del Revenue Net, que será igual al Revenue + Incremento por País
- Si el Country es Colombia, incrementa un 12.5%
- Si el Country es Canadá, incrementa un 8.26%
- Si el Country es USA, incrementa un 48.26%
- Para el resto no haremos incrementos

```python
import pandas as pd

tricolom = trimestre[trimestre["Country"]=="Colombia"]
tricana = trimestre[trimestre["Country"]=="Canada"]
triusa = trimestre[trimestre["Country"]=="USA"]
trialtres = trimestre[~trimestre["Country"].isin(["Colombia", "Canada", "USA"])] #la virgulilla para que me de lo opuesto

tricolom["Revenue Net"] = trimestre["Revenue"] * (1.125)
tricana["Revenue Net"] = trimestre["Revenue"] * (1.0826)
triusa["Revenue Net"] = trimestre["Revenue"] * (1.4826)
trialtres["Revenue Net"] = trimestre["Revenue"]

revnet = pd.concat([tricolom,tricana,triusa,trialtres],ignore_index=True)
revnet 

print("La suma de Revenue Net completa es de",round(revnet["Revenue Net"].sum(),3))
```

El concepto isin, muy importante, se trata de obtener los valores según una lista, ¿puede ser uno sólo? Si, pero igual como lista, con paréntesis y corchetes de todas formas

```python
filtro = trimestre["Country"].isin(["Colombia", "Canada"])
trimestre[filtro]
```

```python
#¿Cómo decir con isin que sea lo contrario de la lista? Con la virgulilla (alt + 126, ~)
filtro = ~trimestre["Country"].isin(["Colombia", "Canada"])
trimestre[filtro]
```

```python
import pandas as pd

pedidos = pd.read_excel("C:\\Users\\formacio\\Downloads\\Pedidos (1).xlsx", sheet_name="Pedidos")
detalles = pd.read_excel("C:\\Users\\formacio\\Downloads\\Detalles Pedidos (1).xlsx", sheet_name="Detalles Pedidos")

#Combinar ambos dataframes en un mismo campo
totalpedidos = pd.merge(pedidos, detalles, on="Id. de pedido", how="inner")
totalpedidos
```

```python
#¿Qué sucedería si los nombres de los campos a coordinar o clave no fueran iguales?
pd.merge(pedidos, detalles, left_on="Id.de pedido", right_on="Num pedido", how="Inner") #no lo pruebes, no funciona porque en este caso si hay comunes
#Debe ser el mismo tipo de dato: número-número, string-string, etc
```

```python
#Vamos a ver agrupaciones
#primero creamops el campo importe = preciu ud x cantidad, queremos ver la suma de importe por País de destinatario
totalpedidos["Importe"] = totalpedidos["Cantidad"] * totalpedidos ["Precio por unidad"]
totalpedidos
#vamos a agrupar
totalpedidos.groupby(["País de destinatario"]).Importe.sum()
```

```python
#La misma agrupación pero con dos dimensiones, sólo por ver que se puede
totalpedidos.groupby(["País de destinatario", "Ciudad de destinatario"]).Importe.sum()
```
